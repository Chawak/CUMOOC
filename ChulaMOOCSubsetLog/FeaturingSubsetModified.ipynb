```python
Q=pd.read_csv("quizlog.csv")
score=pd.read_csv("SCORE.csv")
```

```python
df=pd.read_csv("dfneweraaisus.csv")
df
```

```python
df["UC"]=df["cv_uid"]+df["cv_cid"]
df["quiz_submit"]=0
df["quiz_open"]=0
for i in df["UC"].unique(): 
  x=Q[Q["UC"]==i]["UC"].count()
  for index,row in Q[Q["UC"]==i][:x//8].iterrows():
    if row["action_type"]=="quiz_submitted":
        df.loc[df["UC"]==row["UC"],["quiz_submit"]]+=1
    else:
        df.loc[df["UC"]==row["UC"],["quiz_open"]]+=1
```

```python
df["all_score"]=0
df["nub"]=0
score["avg"]=score["point"]/score["total"]
score["avg"]=score["avg"].fillna(0)

for i in range(df["cv_uid"].count()):
    K=score[(score["cv_uid"]==df["cv_uid"][i]) & (score["TtoC"]==df["cv_cid"][i])]["avg"].count()//8
    df["all_score"][i]+=score[(score["cv_uid"]==df["cv_uid"][i]) & (score["TtoC"]==df["cv_cid"][i])][:K]["avg"].sum()
    df["nub"][i]+=K

df["avg_score"]=(df["all_score"]/df["nub"]).fillna(0)
df.to_csv("df18.csv",index=False)
```
